```python
import pandas as pd
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
for task in [
    "language_detection",
    "named_entity_recognition",
    "object_detection",
    "sentiment_analysis",
]:
    df = pd.read_csv(f"results/{task}.csv")
    sns.boxplot(df, x="n", y="execution_time_s")
    plt.title(task)
    plt.grid(True)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.savefig(f"results/{task}_{datetime.now()}.pdf")
    plt.show()
```
